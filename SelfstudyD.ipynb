# Exercise D.3 (Monte Carlo simulations in numpy)

Recall Exercise B.5 and implement a version of in_unit_circle that uses numpy arrays.
You should take the code from Exercise B.5 as basis and add a new function in_unit_circle_np.

To write this in Python follow these steps:

1. With numpy.random.uniform() create a 2 × N matrix with random numbers.
2. Use element wise computation to get M.
   Hint: True is interpreted as 1 and False as 0 when adding boolean values.

Once you have the second implementation test it and compare it with the other version:

1. Is the accuracy the same?
2. Have a look at timeit and see which of your two code versions is faster for which N.
3. Plot your results for different values of N (execution times for both versions, maybe use a logarithmic axis).
4. Are there other plots that might give you an insight for this example?


```python
import random
import math
import numpy as np
import timeit
import matplotlib.pyplot as plt


def in_unit_circle(N: int) -> int:
    M = 0
    for _ in range(N):
        x = random.random()
        y = random.random()
        if x*x + y*y <= 1:
            M += 1
    return M


def estimate_pi(N: int) -> float:
    return 4 * in_unit_circle(N) / N


def get_accuracy(N: int) -> float:
    return abs(estimate_pi(N) - math.pi)


def in_unit_circle_np(N: int) -> int:
    points = np.random.uniform(0, 1, size=(2, N))
    x = points[0]
    y = points[1]
    inside = (x*x + y*y) <= 1
    return int(inside.sum())


def estimate_pi_np(N: int) -> float:
    return 4 * in_unit_circle_np(N) / N


def get_accuracy_np(N: int) -> float:
    return abs(estimate_pi_np(N) - math.pi)
```

## Compare accuracy for different values of N


```python
Ns = [100, 1_000, 10_000, 100_000]

for N in Ns:
    print("N =", N)
    print("pi (python) =", estimate_pi(N), "accuracy =", get_accuracy(N))
    print("pi (numpy)  =", estimate_pi_np(N), "accuracy =", get_accuracy_np(N))
    print()
```

## Compare execution time using timeit


```python
Ns = [100, 1_000, 10_000, 100_000, 1_000_000]

times_python = []
times_numpy = []

for N in Ns:
    t_py = timeit.timeit(lambda: estimate_pi(N), number=5)
    t_np = timeit.timeit(lambda: estimate_pi_np(N), number=5)
    times_python.append(t_py)
    times_numpy.append(t_np)

times_python, times_numpy
```

## Plot execution times (log scale)


```python
plt.plot(Ns, times_python, label="python loop")
plt.plot(Ns, times_numpy, label="numpy vectorized")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("time (seconds) for 5 runs")
plt.title("Execution time comparison")
plt.legend()
plt.show()
```

## Other plot: accuracy vs N


```python
acc_python = [get_accuracy(N) for N in Ns]
acc_numpy = [get_accuracy_np(N) for N in Ns]

plt.plot(Ns, acc_python, label="python loop")
plt.plot(Ns, acc_numpy, label="numpy vectorized")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("absolute error")
plt.title("Accuracy comparison")
plt.legend()
plt.show()
```

Antwort:

Die Laufzeitmessung zeigt, dass die vektorisierte NumPy-Variante über alle getesteten N schneller ist und der Vorteil mit N wächst.
Die Genauigkeit ist bei beiden Varianten vergleichbar; Unterschiede und Zickzack-Verläufe entstehen durch Zufallsrauschen der Monte-Carlo-Methode.
Für eine stabilere Aussage kann man pro N mehrere Wiederholungen durchführen und Mittelwert sowie Streuung des Fehlers darstellen.


# Exercise D.4 (Predator Prey simulation)

The Lotka-Volterra equations describe a predator-prey system:

ẋ = αx − βxy  
ẏ = δxy − γy

We implement the model using classes for Prey, Predator and a
LotkaVolterra simulation class.


```python
from __future__ import annotations
import numpy as np
import matplotlib.pyplot as plt


class Prey:
    def __init__(self, population: int, natural_growth_rate: float, death_rate_of_prey: float):
        self.population = int(population)                 # public
        self._growth_rate = float(natural_growth_rate)    # private
        self._death_rate = float(death_rate_of_prey)       # private
        self._validate()

    def _validate(self) -> None:
        if self.population < 0:
            raise ValueError("Prey population must be non-negative.")
        if self._growth_rate < 0:
            raise ValueError("Prey growth rate must be non-negative.")
        if self._death_rate < 0:
            raise ValueError("Prey death rate must be non-negative.")

    def getGrowthRate(self) -> float:
        return self._growth_rate

    def setGrowthRate(self, growthRate: float) -> None:
        self._growth_rate = float(growthRate)
        self._validate()

    def getDeathRate(self) -> float:
        return self._death_rate

    def setDeathRate(self, deathRate: float) -> None:
        self._death_rate = float(deathRate)
        self._validate()


class Predator:
    def __init__(self, population: int, growth_rate_per_consumed_prey: float, death_rate_of_predators: float):
        self.population = int(population)                      # public
        self._growth_rate = float(growth_rate_per_consumed_prey)  # private
        self._death_rate = float(death_rate_of_predators)         # private
        self._validate()

    def _validate(self) -> None:
        if self.population < 0:
            raise ValueError("Predator population must be non-negative.")
        if self._growth_rate < 0:
            raise ValueError("Predator growth rate must be non-negative.")
        if self._death_rate < 0:
            raise ValueError("Predator death rate must be non-negative.")

    def getGrowthRate(self) -> float:
        return self._growth_rate

    def setGrowthRate(self, growthRate: float) -> None:
        self._growth_rate = float(growthRate)
        self._validate()

    def getDeathRate(self) -> float:
        return self._death_rate

    def setDeathRate(self, deathRate: float) -> None:
        self._death_rate = float(deathRate)
        self._validate()


class Propagation:
    def __init__(self, prey: Prey, predator: Predator):
        self.prey = prey
        self.predator = predator

    def _calculateDerivates(self, x: float, y: float) -> tuple[float, float]:
        # x' = alpha*x - beta*x*y
        # y' = delta*x*y - gamma*y
        alpha = self.prey.getGrowthRate()
        beta = self.prey.getDeathRate()
        delta = self.predator.getGrowthRate()
        gamma = self.predator.getDeathRate()

        dx = alpha * x - beta * x * y
        dy = delta * x * y - gamma * y
        return dx, dy

    def _propagateTimeStep(self, dt: float) -> None:
        x = float(self.prey.population)
        y = float(self.predator.population)

        dx, dy = self._calculateDerivates(x, y)

        x_next = x + dt * dx
        y_next = y + dt * dy

        # ensure non-negative, keep as float internally but store as float-friendly values
        self.prey.population = max(0.0, x_next)
        self.predator.population = max(0.0, y_next)

    def calcPopOverTime(self, time: float, dt: float) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
        steps = int(time / dt) + 1
        t = np.linspace(0.0, time, steps)

        x = np.zeros(steps)
        y = np.zeros(steps)

        x[0] = self.prey.population
        y[0] = self.predator.population

        for i in range(1, steps):
            self._propagateTimeStep(dt)
            x[i] = self.prey.population
            y[i] = self.predator.population

        return t, x, y
```

## Run the simulation with the given parameters and plot the results


```python
alpha = 1
beta = 0.1
gamma = 0.5
delta = 0.02

x0 = 100
y0 = 20

prey = Prey(population=x0, natural_growth_rate=alpha, death_rate_of_prey=beta)
predator = Predator(population=y0, growth_rate_per_consumed_prey=delta, death_rate_of_predators=gamma)

model = Propagation(prey, predator)

t, x, y = model.calcPopOverTime(time=50, dt=0.01)

plt.plot(t, x, label="prey (x)")
plt.plot(t, y, label="predator (y)")
plt.xlabel("time")
plt.ylabel("population")
plt.title("Lotka-Volterra simulation (Forward Euler)")
plt.legend()
plt.show()
```

Antwort:

Die Simulation zeigt typische periodische Schwankungen:
Wenn die Beutepopulation wächst, steigt zeitverzögert auch die Räuberpopulation.
Durch mehr Räuber sinkt die Beutepopulation wieder, wodurch später auch die Räuberpopulation abnimmt.
Dadurch entstehen wiederkehrende Oszillationen in beiden Populationen.

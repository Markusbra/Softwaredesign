## B.2 Lists, conditionals and loops

### For two given lists list1 and list2 of integers with equal length define the following new lists:

1. list_zip which combines the two lists to a single list like a zip. So start with the first element of list1, followed by the first element of list2, than the second elements until all elements of the lists are combined.

```python
list1 = [1, 2, 3, 4, 5]
list2 = [11, 12, 13, 14, 15]

# 1. list_zip
list_zip = []
for i in range(len(list1)):
    list_zip.append(list1[i])
    list_zip.append(list2[i])

print("list_zip:", list_zip)
```

2. list_odd which contains only the odd integers of list1, followed by the odd integers of list2.

```python
# 2. list_odd
list_odd = []

for x in list1:
    if x % 2 != 0:
        list_odd.append(x)

for x in list2:
    if x % 2 != 0:
        list_odd.append(x)

print("list_odd:", list_odd)
```

3. list_zip_reverse which combines the elements of the lists like for list_zip but starts at the last element of list1.

```python
# 3. list_zip_reverse
list_zip_reverse = []
for i in range(len(list1) - 1, -1, -1):
    list_zip_reverse.append(list1[i])
    list_zip_reverse.append(list2[i])

print("list_zip_reverse:", list_zip_reverse)
```

## Exercise B.5 (Monte Carlo simulations with random numbers, functions, conditionals and loops)

### Random points are generated in the first quadrant and checked whether they lie inside the unit circle.
The value of π is estimated from the ratio of points inside the circle to the total number of points.

### Steps (as required)

1. Find a Python module that can generate random values.
2. Define a function in_unit_circle(N) that computes and returns M, the number of points (out of N) that lie inside the unit circle in the first quadrant.
   - Hint: Interpret two random numbers between 0 and 1 as Cartesian coordinates (x, y).  
     The point is inside the unit circle if x^2 + y^2 <= 1.
3. Define a function estimate_pi(N) that computes and returns an approximation of π using Equation B.1:
   
$$
\pi \approx 4 \cdot \frac{M}{N}
$$

4. Find a Python module that provides the exact value of π and define a function get_accuracy(N) that returns the absolute difference between the estimate and the exact value.
5. Test your functions for different values of N (note: N must be large to obtain multiple correct digits of π).



```python
import random

def in_unit_circle(N):
    M = 0
    
    for _ in range(N):
        x = random.random()  # Zahl zwischen 0 und 1
        y = random.random()
        
        if x**2 + y**2 <= 1:
            M += 1
            
    return M

def estimate_pi(N):
    M = in_unit_circle(N)
    return 4 * M / N

import math

def get_accuracy(N):
    return abs(estimate_pi(N) - math.pi)

for N in [100, 1000, 10000, 100000]:
    print("N =", N)
    print("π Schätzung:", estimate_pi(N))
    print("Fehler:", get_accuracy(N))
    print()
```

### Hinweis auf positive ganze Zahlen

Mit Type Hints kann in Python nur der Datentyp int angegeben werden, aber keine zusätzliche Einschränkungen wie „positiv“.

Um anzugeben, dass die Zahl positiv sein soll, kann man:

- das in der Dokumentation beschreiben oder

```python
def in_unit_circle(N: int) -> int:
    
    if N <= 0:
        raise ValueError("N must be a positive integer")
    
    M = 0
    
    for _ in range(N):
        x = random.random()
        y = random.random()
        
        if x**2 + y**2 <= 1:
            M += 1
            
    return M
```

## Exercise B.9 — Recursion with Fibonacci

Create a function that computes the n-th Fibonacci number using recursion,
i.e., a function that calls itself.

The Fibonacci sequence is defined as:

- x₀ = 0
- x₁ = 1
- xₙ = xₙ₋₁ + xₙ₋₂ for n ≥ 2

The first elements of the sequence are:
0, 1, 1, 2, 3, 5, 8, 13, 21, ...


```python
def fibonacci(n: int) -> int:
    
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(10):
    print(f"F({i}) =", fibonacci(i))
```

### Antwort

Die Funktion fibonacci(n) berechnet rekursiv die n-te Fibonacci-Zahl.
Für n = 0 und n = 1 werden feste Werte zurückgegeben.
Für größere Werte ruft sich die Funktion selbst mit n-1 und n-2 auf und addiert die Ergebnisse.


## Exercise B.10 — Working with dictionaries

Compute the Arabic number from a Roman numeral for positive integers.

Use the provided dictionary to map Roman symbols to their values.


```python
roman = {
    "I": 1, "IV": 4, "V": 5, "IX": 9,
    "X": 10, "XL": 40, "L": 50, "XC": 90,
    "C": 100, "CD": 400, "D": 500,
    "CM": 900, "M": 1000
}

def roman_to_arabic(s: str) -> int:
    total = 0
    i = 0
    
    while i < len(s):
        
        # check two-character symbols (e.g. IV, IX, XL, ...)
        if i + 1 < len(s) and s[i:i+2] in roman:
            total += roman[s[i:i+2]]
            i += 2
        else:
            total += roman[s[i]]
            i += 1
            
    return total

print(roman_to_arabic("MMMDCCXXIV"))
print(roman_to_arabic("MXCIV"))
print(roman_to_arabic("VIII"))
```

### Antwort

Die Funktion liest das römische Zahlzeichen von links nach rechts, wobei geprüft wird, ob ein oder zwei Zeichen zusammen einen Wert im Dictionary bilden.
Der entsprechende Wert wird zur Summe addiert, bis die gesamte Zeichenkette verarbeitet ist.

```python
def words_by_length(sentence: str) -> dict[int, set[str]]:
    ws = sentence.split()
    d: dict[int, set[str]] = {}
    for w in ws:
        d.setdefault(len(w), set()).add(w)
    return d

tests = [
    "I spy with my little eye a tricky list and dict comprehension coming up",
    "this is another short sentence for testing",
]

for s in tests:
    print(words_by_length(s))
```

```python
def words_by_length_comp(sentence: str) -> dict[int, set[str]]:
    ws = sentence.split()
    return {L: {w for w in ws if len(w) == L} for L in set(map(len, ws))}

words_by_length_comp("I spy with my little eye a tricky list and dict comprehension coming up")
```
